#### Exercise 1: Solar Energy Calculation

<p style="text-align:center">
    <img src="https://th.bing.com/th/id/R.b3c79851427a8a55723fe3827b31abec?rik=hhIrP8ict8gxIw&riu=http%3a%2f%2fnews.delaware.gov%2ffiles%2f2012%2f03%2fMotech-Solar-Panel-floating.jpg&ehk=Q3uAu91jOrnoGRCY6%2bd9acDMq6CEzBIKGJLdw9zD5B0%3d&risl=&pid=ImgRaw&r=0" width="500" height="200" alt="Solar Energy"  />
</p>

$$
E = A \times \eta \times I \times t
$$


**This exercise is designed to help you understand how to use `variables` and perform basic `arithmetic operations` in `Python`.**

**The task is to calculate the `amount of solar energy that can be generated by a solar panel of a given size in a day`.**

**You will need to use the formula:

`Energy = Solar Panel Area * Solar Panel Efficiency * Solar Irradiance * Time`.**

Assume the given values:
- `Panel Area` (1.2 m^2)
- `Efficiency` (20%)
- `Irradiance` (1000W/m^2)
- `Time` (12 hours).

#### Exercise 2: Wind Energy Production

<p style="text-align:center">
    <img src="https://www.greenbookmark.com/sites/default/files/images/wind-energy-production-1.jpg" width="500" height="200" alt="Wind energy"  />
</p>

$$
P = 0.5 \times \rho \times A \times V^3
$$


**This exercise introduces you to `control structures` in `Python`, such as `if-else` statements.**

**The task is to simulate the operation of a wind turbine based on the wind speed.**

**You will need to create conditions for when `the turbine is off (when the wind speed is less than 3 m/s or above 25 m/s)` and `when it’s on (when the wind speed is between 3 m/s and 25 m/s)`.**

**When the turbine is on, the program should `calculate the power produced using the given formula`:**

**Power = 0.5 * Air Density * Area * (Wind Speed)^3.**

Assume the given values:

- A `diamete`r of 120 meters.
- The `air density` is 1.225 kg/m^3.
- The `wind speed` is 10 m/s.

Recall that:`Area = π * (Diameter / 2)^2`

#### Exercise 3: Electric Vehicle Charging Time Calculation

<p style="text-align:center">
    <img src="https://queenstownsolar.co.nz/app/uploads/EV-Charging.jpg" width="500" height="200" alt="Wind energy"  />
</p>

$$
\text{Time} = \frac{\text{Battery Capacity}}{\text{Charging Power}}
$$

**This exercise introduces you to the concept of `functions in Python`.**

**The task is to simulate the charging process of an electric vehicle.**

**You will need to create a `function` that takes the `battery capacity` and `charging power` as inputs, and returns the `charging time`.**

Assume the given values:

- `Battery Capacity`: 100 kWh
- `Charging Power`: 50 kW

#### Exercise 4: Energy Consumption Calculation

$$
\text{Total Energy} = \sum_{i=1}^{n} \text{Daily Energy}_i
$$


**This exercise introduces you to the concept of `loops`, `user inputs`, and `type casting` in `Python`.**

**The task is to create a program that calculates the `total energy consumption of an electric vehicle over a certain number of days`.**

**The program should ask the user for the `number of days` and the `daily energy consumption`, and then calculate the `total energy consumption`.**
